```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
dataSet="Dataset/Ecommerce Customers.csv"
```

```python
df=pd.read_csv(dataSet)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
new_df=df.copy()
```

```python
new_df.drop(['Email','Address','Avatar'],axis=1,inplace=True)
```

```python
sns.pairplot(new_df, size=5)
```

```python
x=np.array(new_df.drop(['Yearly Amount Spent'],axis=1))
y=np.array(new_df['Yearly Amount Spent'])
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=450)
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
len(x_test)
```

```python
len(y_test)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.scatter(y_test,y_pred)
plt.grid()
plt.plot([min(y_test),max(y_test)],[min(y_pred),max(y_pred)],color='red',linewidth=2)
```
